```python
import hashlib
import base64
import itertools


# =========================================================
#               MERCURY CIPHER
# =========================================================

TEXT_TO_MERCURY = {
    'a': '☿',
    'b': '☽',
    'c': '☼',
    'd': '♁',
    'e': '♂',
    'f': '♀',
    'g': '♃',
    'h': '♄',
    'i': '🜁',
    'j': '🜂',
    'k': '🜃',
    'l': '🜄',
    'm': '⚕',
    'n': '⚚',
    'o': '⚛',
    'p': '⚜',
    'q': '⚝',
    'r': '⚡',
    's': '☢',
    't': '☣',
    'u': '☤',
    'v': '☥',
    'w': '☦',
    'x': '☧',
    'y': '☨',
    'z': '☩',

    '0': '⓿',
    '1': '❶',
    '2': '❷',
    '3': '❸',
    '4': '❹',
    '5': '❺',
    '6': '❻',
    '7': '❼',
    '8': '❽',
    '9': '❾',

    ' ': '/',
    '.': '•',
    ',': '‚',
    '?': '¿',
    '!': '¡'
}

# Reverse Mapping
MERCURY_TO_TEXT = {
    v: k for k, v in TEXT_TO_MERCURY.items()
}


# =========================================================
#               ENCODE TO MERCURY CODE
# =========================================================

def text_to_mercury(text):

    encoded = []

    for char in text.lower():

        if char in TEXT_TO_MERCURY:
            encoded.append(TEXT_TO_MERCURY[char])

        else:
            encoded.append(char)

    return " ".join(encoded)


# =========================================================
#               DECODE MERCURY CODE
# =========================================================

def mercury_to_text(cipher):

    decoded = []

    for symbol in cipher.split():

        if symbol in MERCURY_TO_TEXT:
            decoded.append(MERCURY_TO_TEXT[symbol])

        else:
            decoded.append(symbol)

    return "".join(decoded)


# =========================================================
#               PASSWORD VALIDATION
# =========================================================

def validate_password(password):

    special_chars = "!@#$%^&*()-_=+[]{}|;:'\",.<>?/`~"

    if len(password) != 8:
        return False

    has_upper = any(c.isupper() for c in password)
    has_lower = any(c.islower() for c in password)
    has_digit = any(c.isdigit() for c in password)
    has_special = any(c in special_chars for c in password)

    return (
        has_upper and
        has_lower and
        has_digit and
        has_special
    )


# =========================================================
#               KEY GENERATION
# =========================================================

def generate_key(password):

    return hashlib.sha256(
        password.encode()
    ).digest()


# =========================================================
#               XOR ENCRYPTION
# =========================================================

def xor_encrypt(data, key):

    data_bytes = data.encode()

    encrypted = bytes(
        a ^ b for a, b in zip(
            data_bytes,
            itertools.cycle(key)
        )
    )

    return base64.b64encode(
        encrypted
    ).decode()


# =========================================================
#               XOR DECRYPTION
# =========================================================

def xor_decrypt(data, key):

    encrypted_bytes = base64.b64decode(data)

    decrypted = bytes(
        a ^ b for a, b in zip(
            encrypted_bytes,
            itertools.cycle(key)
        )
    )

    return decrypted.decode()


# =========================================================
#               MAIN PROGRAM
# =========================================================

def main():

    print("=======================================")
    print("     MERCURY CODE ENCRYPTION SYSTEM")
    print("=======================================")

    # ----------------------------------------------------
    # INPUT
    # ----------------------------------------------------
    original_text = input("\nEnter text: ").strip()

    # ----------------------------------------------------
    # MERCURY ENCODE
    # ----------------------------------------------------
    mercury_cipher = text_to_mercury(original_text)

    print("\n☿ Mercury Cipher:")
    print(mercury_cipher)

    # ----------------------------------------------------
    # PASSWORD
    # ----------------------------------------------------
    password = input(
        "\nEnter 8-character password "
        "(Upper + Lower + Digit + Special): "
    ).strip()

    # ----------------------------------------------------
    # VALIDATE PASSWORD
    # ----------------------------------------------------
    if not validate_password(password):

        print("\n❌ Invalid Password!")

        print("Password must contain:")
        print("✔ 8 characters")
        print("✔ Uppercase")
        print("✔ Lowercase")
        print("✔ Digit")
        print("✔ Special character")

        return

    # ----------------------------------------------------
    # GENERATE KEY
    # ----------------------------------------------------
    key = generate_key(password)

    # ----------------------------------------------------
    # ENCRYPT
    # ----------------------------------------------------
    encrypted_output = xor_encrypt(
        mercury_cipher,
        key
    )

    print("\n🔐 Encrypted Mercury Cipher:")
    print(encrypted_output)

    # ----------------------------------------------------
    # DECRYPT OPTION
    # ----------------------------------------------------
    choice = input(
        "\nDo you want to decrypt? (yes/no): "
    ).strip().lower()

    if choice == "yes":

        decrypt_password = input(
            "\nEnter same password: "
        ).strip()

        if decrypt_password != password:

            print("\n❌ Incorrect Password!")
            return

        try:

            decrypt_key = generate_key(
                decrypt_password
            )

            decrypted_mercury = xor_decrypt(
                encrypted_output,
                decrypt_key
            )

            print("\n☿ Decrypted Mercury Cipher:")
            print(decrypted_mercury)

            original_output = mercury_to_text(
                decrypted_mercury
            )

            print("\n🔓 Original Text:")
            print(original_output)

        except Exception as e:

            print("\n❌ Decryption Failed!")
            print("Error:", str(e))

    else:

        print("\n👋 Good luck!")


# =========================================================
#               RUN PROGRAM
# =========================================================

if __name__ == "__main__":
    main()
```
